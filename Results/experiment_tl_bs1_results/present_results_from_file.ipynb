```python
import pickle
import matplotlib.pyplot as plt
```

```python
for learning_rate in [1, 0.9, 0.3, 0.09, 0.03, 0.009, 0.003, 0.0009, 0.0003, 9e-5]:
    with open(str(learning_rate) + '.pickle', 'rb') as f:
        train_losses, val_losses, train_accuracies, val_accuracies = pickle.load(f)

        plt.figure()
        plt.subplot(311)
        plt.title('Training loss')
        plt.plot(train_losses, 'o')
        plt.xlabel('Iteration')

        plt.subplot(312)
        plt.title('Validation loss')
        plt.plot(val_losses, 'o')
        plt.xlabel('Iteration')

        plt.subplot(313)
        plt.title('Accuracy')
        plt.plot(train_accuracies, '-o', label='Train Accuracy')
        plt.plot(val_accuracies, '-o', label='Val Accuracy')
        plt.xlabel('Iteration')
        plt.legend(loc='lower center')

        plt.show()
        plt.savefig('transfer_learning'+str(learning_rate)+'.pdf')
```

```python
with open('0.09.pickle', 'rb') as f:
    train_losses, val_losses, train_accuracies, val_accuracies = pickle.load(f)
    print("Train Accuracy: {}, Val Accuracy: {}".format(train_accuracies[-1], val_accuracies[-1]))
```
